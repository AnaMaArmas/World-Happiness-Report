▪ What is the average happiness score across regions? Does it change over time? Is it on average 
increasing or decreasing?  In the 5 years, what countries increased their happiness scores? Which ones decreased their 
happiness score? 

```python
# Group the data by the 'Region' column and calculate the mean happiness score for each year
region_score15 = df_2015_cleaned.groupby('Region')['Happiness Score'].mean().reset_index(name='mean2015').sort_values(['mean2015'], ascending=False)
region_score16 = df_2016_cleaned.groupby('Region')['Happiness Score'].mean().reset_index(name='mean2016').sort_values(['mean2016'], ascending=False)
region_score17 = df_2017_cleaned.groupby('Region')['Happiness Score'].mean().reset_index(name='mean2017').sort_values(['mean2017'], ascending=False)
region_score18 = df_2018_cleaned.groupby('Region')['Happiness Score'].mean().reset_index(name='mean2018').sort_values(['mean2018'], ascending=False)
region_score19 = df_2019_cleaned.groupby('Region')['Happiness Score'].mean().reset_index(name='mean2019').sort_values(['mean2019'], ascending=False)

# Merge the dataframes to create a single dataframe containing the mean scores for each region
region_score = region_score15.copy()
region_score = region_score.merge(region_score16, on=['Region'])
region_score = region_score.merge(region_score17, on=['Region'])
region_score = region_score.merge(region_score18, on=['Region'])
region_score = region_score.merge(region_score19, on=['Region'])

# Calculate the difference in mean scores between 2015 and 2019
region_score['diff mean 2015-2019'] = region_score['mean2019'] - region_score['mean2015']

region_score
```

```python
# Create subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(14, 9))

# Plot the bar chart for 2015 Happiness Score
ax1.bar(region_score['Region'], region_score['mean2015'], color='navy', linewidth=3)
ax1.set_title('2015 Happiness Score', fontsize=20)
ax1.set_ylabel('Happiness Score', fontsize=16)
ax1.yaxis.set_tick_params(labelsize=14)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90)
ax1.xaxis.set_tick_params(labelsize=14)
ax1.grid(True)

# Plot the bar chart for 2019 Happiness Score
ax2.bar(region_score['Region'], region_score['mean2019'], color='steelblue', linewidth=3)
ax2.set_title('2019 Happiness Score', fontsize=20)
ax2.set_ylabel('Happiness Score', fontsize=16)
ax2.yaxis.set_tick_params(labelsize=14)
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=90)
ax2.xaxis.set_tick_params(labelsize=14)
ax2.grid(True)

# Add a title to the entire figure
fig.suptitle("Average Happiness Scores in Different World Regions", fontsize=28)

# Adjust layout
plt.tight_layout()
```

The average Happiness Score stays similar in the 5 years. There were only a few regions that improved their overall happiness score, both of the European regions improved their score. And Latin America and the Caribbean decreased their happiness score the most.

By comparing the yearly global average Healthy Life Expectancy with the Averga GDP per Capita we uncovered further insights:
1. The mirrored relationship indicates that both variables follow the same trend. It might be a negative correlation since one line is going upwards and one downwards. I suggest further investigation to understand this correlation.
2. A negative correlation between GDP and Life Expectancy seems counterintuitive at first glance but there might be some explanations:
- Economic Development vs. Health Infrastructure: in some cases, rapid economic development might lead to environmental degradation, an increase in stress, and unhealthy lifestyles, which could negatively impact health perception.
- Prioritization of Resources: Sometimes, countries prioritize economic growth over investments in healthcare and social services. This prioritization might lead to improvements in GDP but could result in neglected healthcare infrastructure, leading to poorer health outcomes.
- Time Lag: Changes in economic conditions might take time to translate into improvements or deteriorations in health outcomes. Thus, short-term fluctuations in economic growth might not immediately reflect corresponding changes in health indicators.
- Data and Measurement: Since this data was collected via survey the quality and accuracy of the data used to measure both GDP per Capita and Healthy Life Expectancy is based on personal perception. This limitation in the data collection method can influence the correlations.
