```python
# Create empty sets to store unique country lists
country_sets = []

# Iterate over each DataFrame
for df in [df_2015, df_2016, df_2017, df_2018, df_2019]:
    # Print the columns of the DataFrame for debugging
    print("Columns:", df.columns)
    
    # Check if the column name is 'Country'
    if 'Country' in df.columns:
        country_column_name = 'Country'
    elif 'Country or region' in df.columns:  # Check for alternate column name
        country_column_name = 'Country or region'
    else:
        raise ValueError("Country column not found in DataFrame")
    
    # Get unique countries for each DataFrame
    country_set = set(df[country_column_name])
    
    # Add to the list of sets
    country_sets.append(country_set)

# Initialize a dictionary to store missing or different countries
diff_countries = {}

# Iterate over pairs of consecutive DataFrames
for i in range(len(country_sets) - 1):
    # Find the difference between consecutive sets
    diff = country_sets[i] - country_sets[i+1]
    
    # Store the difference in the dictionary
    diff_countries[f'{i+2015}-{i+1+2015}'] = diff

# Print the results
for key, value in diff_countries.items():
    print(f"Difference between {key}: {len(value)} countries")
    print(value)
```

# Data Wrangling df_2017

```python
df_2017.head(15)
```

```python
#Dropping columns
df_2017 = df_2017.drop(['Whisker.high', 'Whisker.low', 'Dystopia.Residual'], axis=1)
```

```python
#renaming columns 
df_2017 = df_2017.rename(columns={
    'Happiness.Rank': 'Happiness Rank',
    'Happiness.Score': 'Happiness Score',
    'Economy..GDP.per.Capita.': 'GDP per Capita',
    'Family': 'Social Support',
    'Health..Life.Expectancy.': 'Healthy Life Expectancy',
    'Freedom': 'Freedom to Make Life Choices',
    'Trust..Government.Corruption.': 'Perceptions of Corruption'
})
```

```python
df_2017.head(15)
```

```python
# Group the data by the 'Region' column in df_2016 and aggregate the 'Country' column
countries_by_region = df_2016.groupby('Region')['Country'].apply(list)

# Create a mapping dictionary from country to region
country_to_region = {}
for region, countries in countries_by_region.items():
    for country in countries:
        country_to_region[country] = region

# Add additional countries and their assigned regions
additional_countries = {
    'Taiwan Province of China': 'Asia',
    'Hong Kong S.A.R., China': 'Asia',
    'Mozambique': 'Africa',
    'Lesotho': 'Africa',
    'Central African Republic': 'Africa'
}
country_to_region.update(additional_countries)

# Add the 'Region' column to df_2017 using the updated mapping
df_2017['Region'] = df_2017['Country'].map(country_to_region)
```

```python
# Filter the DataFrame to select rows where the 'Region' column is missing
countries_without_region = df_2017[df_2017['Region'].isnull()]['Country']

# Print the countries without a region assigned
print("Countries without a region assigned in df_2017:")
print(countries_without_region)
```
